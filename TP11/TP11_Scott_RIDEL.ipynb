## Question 4
Interrogez VOTRE serveur de bases de données pour afficher la liste des villes et leurs populations respectives pour le pays d’Europe ayant la population la plus nombreuse. Vous pourrez pour cela utiliser un DataFrame du module pandas.

```python
import mysql.connector

config = {
  'user': 'root',
  'password': '',
  'host': '127.0.0.1',
  'database': 'world',
  'raise_on_warnings': True
}


cnx = mysql.connector.connect(**config)
curseur = cnx.cursor()

query = ("SELECT Name FROM country WHERE Continent = 'Europe' ORDER BY Population DESC LIMIT 1")
curseur.execute(query)

for Name in curseur:
    curseur2 = cnx.cursor()
    query2 = (f"SELECT city.Name, city.Population FROM city, country WHERE city.CountryCode = country.Code AND country.Name = '{Name[0]}' ORDER BY city.Population DESC")
    curseur2.execute(query2)
              
    for Name, Population in curseur2:
        print(f"{Name}, {Population}")

curseur2.close()
curseur.close()
cnx.close()
```

## Question 5
Ecrire une fonction Python qui retourne la liste des continents de la base de données world

```python
import mysql.connector

config = {
  'user': 'root',
  'password': '',
  'host': '127.0.0.1',
  'database': 'world',
  'raise_on_warnings': True
}

cnx = mysql.connector.connect(**config)

def continent(cnx):
    curseur = cnx.cursor()
    query = ("SELECT DISTINCT Continent FROM country")
    curseur.execute(query)
    for Continent in curseur:
        print(f"{Continent[0]}")
    curseur.close()
    
continent(cnx)
cnx.close()
```

## Question 7
Ecrire une fonction Python qui accepte un argument nommé continent et sauve pour chaque pays l’ensemble des informations relatives à ces pays, suivi de la liste des villes de chaquepays et de leurs populations respectives dans un format HTML (vous pourrez, pour cela, utiliser la méthode tohtml du module pandas.

```python
import mysql.connector
import pandas as pd

config = {
  'user': 'root',
  'password': '',
  'host': '127.0.0.1',
  'database': 'world',
  'raise_on_warnings': True
}

cnx = mysql.connector.connect(**config)

def body(df):
    body = df.to_html()
    return body.replace('<table border="1" class="dataframe">','<table>')

def myQyery(cnx, query):
    curseur = cnx.cursor()
    curseur.execute(query)
    df = pd.DataFrame(curseur.fetchall(), columns=curseur.column_names)
    curseur.close()
    return df
    
def get_file(root):
    file = open(root, "r")
    r = file.read()
    file.close()
    return r

def city_continent(cnx, continent):
    L = {}
    df1 = myQyery(cnx, f"SELECT * FROM country WHERE Continent = '{continent}'").set_index('Name')
    df2 = myQyery(cnx, f"SELECT country.Name AS 'cn', city.Name, city.Population FROM city, country WHERE city.CountryCode = country.Code AND country.Continent = '{continent}'")
    
    # On liste chaque villes de chaque pays 
    for x in df1.index:
        for y in range(df2.shape[0]):
            if df2['cn'][y] == x:
                if x in L: 
                    L[x] += str(df2['Name'][y])+"(" + str(df2['Population'][y]) + "), "
                else:
                    L[x] = str(df2['Name'][y])+"(" + str(df2['Population'][y]) + "), "
           
    df3 = pd.DataFrame.from_dict(L, orient = 'index')
    
    # Construction de la page HTML
    title1 = f"<h2>Informations sur les pays d'{continent}</h2>"
    title2 = f"<h2>Liste des villes et leur population pour chaque pays d'{continent}</h2>"
    text_file = open("index.html", "w")
    text_file.write(get_file("header.html") + title1 + body(df1) + title2 + body(df3) + get_file("footer.html"))
    text_file.close()
    
# Début du programme
city_continent(cnx, "Europe")
cnx.close()
print("done")
```
